```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Chuẩn hóa dữ liệu GT
df['GT'] = df['GT'].astype(str).str.strip()

# Lọc học sinh nam
df_nam = df[df['GT'].isin(['Nam', '1', 'M'])]

# Kiểm tra dữ liệu
if df_nam.empty:
    print("Không có dữ liệu học sinh nam sau khi lọc")
else:
    tan_so_dt = df_nam['DT'].value_counts()
    tan_suat_dt = df_nam['DT'].value_counts(normalize=True) * 100

    bang_dt_nam = pd.DataFrame({
        'Tần số': tan_so_dt,
        'Tần suất (%)': tan_suat_dt.round(2)
    })

    display(bang_dt_nam)

    # Biểu đồ cột
    plt.figure()
    tan_so_dt.plot(kind='bar')
    plt.title("Biểu đồ tần số dân tộc (DT) của học sinh nam")
    plt.xlabel("Dân tộc")
    plt.ylabel("Tần số")
    plt.show()

    # Biểu đồ tròn
    plt.figure()
    tan_suat_dt.plot(kind='pie', autopct='%1.1f%%')
    plt.title("Biểu đồ tần suất dân tộc (DT) của học sinh nam")
    plt.ylabel("")
    plt.show()
```

```python
#Cau4
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df.head()


df['GT'] = df['GT'].astype(str).str.strip()
df['DT'] = df['DT'].astype(str).str.strip()
df['KV'] = df['KV'].astype(str).str.strip()

df_loc = df[
    (df['GT'].isin(['Nam', '1', 'M'])) &   
    (df['DT'] == 'Kinh') &                 
    (df['DH1'] >= 5.0) &
    (df['DH2'] >= 4.0) &
    (df['DH3'] >= 4.0)
]

df_loc.head()
df_loc.shape

if df_loc.empty:
    print(" Không có học sinh nam dân tộc Kinh thỏa điều kiện điểm")
else:
    tan_so_kv = df_loc['KV'].value_counts()
    tan_suat_kv = df_loc['KV'].value_counts(normalize=True) * 100

    bang_kv = pd.DataFrame({
        'Tần số': tan_so_kv,
        'Tần suất (%)': tan_suat_kv.round(2)
    })

    display(bang_kv)

if not df_loc.empty:
    plt.figure()
    tan_so_kv.plot(kind='bar')
    plt.title("Biểu đồ tần số khu vực (KV)\n(Nam – Kinh – đạt điều kiện DH)")
    plt.xlabel("Khu vực")
    plt.ylabel("Tần số")
    plt.show()

if not df_loc.empty:
    plt.figure()
    tan_suat_kv.plot(kind='pie', autopct='%1.1f%%')
    plt.title("Biểu đồ tần suất khu vực (KV)\n(Nam – Kinh – đạt điều kiện DH)")
    plt.ylabel("")
    plt.show()
```

```python
#Cau5
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")


df['KV'] = df['KV'].astype(str).str.strip()


df_2nt = df[df['KV'] == '2NT']


bien_dh = ['DH1', 'DH2', 'DH3']


for bien in bien_dh:
    print(f"\n================ {bien} =================")

    
    data = df_2nt[df_2nt[bien] >= 5.0][bien]

    if data.empty:
        print(f"⚠️ Không có dữ liệu thỏa điều kiện cho {bien}")
        continue

    
    print("Thống kê mô tả:")
    print(data.describe())

    
    plt.figure()
    plt.hist(data)
    plt.title(f"Histogram {bien} (KV = 2NT, ≥ 5.0)")
    plt.xlabel("Điểm")
    plt.ylabel("Tần số")
    plt.show()

   
    plt.figure()
    plt.boxplot(data)
    plt.title(f"Boxplot {bien} (KV = 2NT, ≥ 5.0)")
    plt.ylabel("Điểm")
    plt.show()
```

```python
#Cau1 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

df['GT'] = df['GT'].astype(str).str.strip()
df_nu = df[df['GT'].isin(['Nu', 'Nữ', '0', 'F'])]

nhom_xl = ['XL1', 'XL2', 'XL3']
muc_xl = ['Y', 'TB', 'K', 'G', 'XS']

bang_dem = pd.DataFrame(index=muc_xl)

for xl in nhom_xl:
    bang_dem[xl] = df_nu[xl].value_counts().reindex(muc_xl, fill_value=0)

bang_dem
x = np.arange(len(nhom_xl))        
width = 0.15                       

plt.figure(figsize=(8, 5))

for i, muc in enumerate(muc_xl):
    plt.bar(
        x + i * width,
        bang_dem.loc[muc],
        width=width,
        label=muc
    )

plt.xticks(x + width * 2, nhom_xl)
plt.xlabel("Nhóm xếp loại")
plt.ylabel("Số lượng học sinh nữ")
plt.title("Trực quan học sinh nữ theo XL1, XL2, XL3 (Unstacked)")
plt.legend(title="Xếp loại")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")


df["KT"] = df["KT"].astype(str).str.strip()
df["KV"] = df["KV"].astype(str).str.strip()
df["KQXT"] = df["KQXT"].astype(str).str.strip()


df_filter = df[
    (df["KT"].isin(["A", "A1", "B"])) &
    (df["KV"].isin(["1", "2", "KV1", "KV2"]))
]

print("Số dòng sau khi lọc:", len(df_filter))


if df_filter.empty:
    print(" Không có dữ liệu thỏa điều kiện!")
else:
    kqxt_count = df_filter["KQXT"].value_counts().sort_index()

    print("\nBảng tần số KQXT:")
    print(kqxt_count)

    
    plt.figure(figsize=(8, 5))
    kqxt_count.plot(kind="bar")

    plt.title("Phân bố KQXT của HS khối A, A1, B (KV 1 & 2)")
    plt.xlabel("Kết quả xét tuyển (KQXT)")
    plt.ylabel("Số lượng học sinh")
    plt.xticks(rotation=0)
    plt.grid(axis="y", linestyle="--", alpha=0.6)

    plt.tight_layout()
    plt.show()
```
