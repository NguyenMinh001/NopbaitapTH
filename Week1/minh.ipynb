```python
# Cau 6
import pandas as pd

df = pd.read_csv("dulieuxettuyendaihoc.csv")


cot_diem = df.select_dtypes(include=["int64", "float64"]).columns


print("Thiếu trước khi xử lý:")
print(df[cot_diem].isna().sum())


for col in cot_diem:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)


print("Thiếu sau khi xử lý:")
print(df[cot_diem].isna().sum())
```

```python
# Cau 8
import pandas as pd

df = pd.read_csv("dulieuxettuyendaihoc.csv")
print(df.head(10))

bins = [0, 5, 6.5, 8, 9, 10]
labels = ["Y", "TB", "K", "G", "XS"]

df["TBM1"] = (
    df["T1"]*2 + df["L1"] + df["H1"] + df["S1"]
    + df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]
) / 10

df["TBM2"] = (
    df["T2"]*2 + df["L2"] + df["H2"] + df["S2"]
    + df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]
) / 10

df["TBM3"] = (
    df["T3"]*2 + df["L3"] + df["H3"] + df["S3"]
    + df["V3"]*2 + df["X3"] + df["D3"] + df["N3"]
) / 10

df["XL1"] = pd.cut(df["TBM1"], bins=bins, labels=labels, right=False)
df["XL2"] = pd.cut(df["TBM2"], bins=bins, labels=labels, right=False)
df["XL3"] = pd.cut(df["TBM3"], bins=bins, labels=labels, right=False)

def xep_loai(tbm):
    if tbm < 5:
        return "Y"
    elif tbm < 6.5:
        return "TB"
    elif tbm < 8:
        return "K"
    elif tbm < 9:
        return "G"
    else:
        return "XS"

df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)

print(df[["TBM1", "XL1", "TBM2", "XL2", "TBM3", "XL3"]].head())
```

```python
# Cau9
import pandas as pd

df = pd.read_csv("dulieuxettuyendaihoc.csv")
print(df.head(10))

df["TBM1"] = (
    df["T1"]*2 + df["L1"] + df["H1"] + df["S1"]
    + df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]
) / 10

df["TBM2"] = (
    df["T2"]*2 + df["L2"] + df["H2"] + df["S2"]
    + df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]
) / 10

df["TBM3"] = (
    df["T3"]*2 + df["L3"] + df["H3"] + df["S3"]
    + df["V3"]*2 + df["X3"] + df["D3"] + df["N3"]
) / 10

df["US_TBM1"] = df["TBM1"] * 0.4
df["US_TBM2"] = df["TBM2"] * 0.4
df["US_TBM3"] = df["TBM3"] * 0.4

df["US_TBM1"] = df["US_TBM1"].round(2)
df["US_TBM2"] = df["US_TBM2"].round(2)
df["US_TBM3"] = df["US_TBM3"].round(2)

print(df[["TBM1","US_TBM1","TBM2","US_TBM2","TBM3","US_TBM3"]].head())

df.to_csv("processed_dulieuxettuyendaihoc.csv", index=False)
```

```python
# Cau 10
import pandas as pd

df = pd.read_csv("dulieuxettuyendaihoc.csv")
print(df.head(10))

def ket_qua_xet_tuyen(row):
    if row["KT"] in ["A", "A1"]:
        diem = (row["DH1"]*2 + row["DH2"] + row["DH3"]) / 4
    elif row["KT"] == "B":
        diem = (row["DH1"] + row["DH2"]*2 + row["DH3"]) / 4
    else:
        diem = (row["DH1"] + row["DH2"] + row["DH3"]) / 3

    if diem >= 5.0:
        return 1  
    else:
        return 0  

df["KQXT"] = df.apply(ket_qua_xet_tuyen, axis=1)

print(df[["KT","DH1","DH2","DH3","KQXT"]].head(10))
```
